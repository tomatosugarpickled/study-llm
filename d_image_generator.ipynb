### Image Generator

| 항목 | gpt-image-1.5 | gpt-image-1-mini |
| :--- | :--- | :--- |
| **품질** | 매우 높음 | 보통 |
| **속도** | 느린 편 | 매우 빠름 |
| **비용** | 상대적으로 비쌈 | 저렴 |

```python
import os, base64
from datetime import datetime


def generate_image_from_text(content):
    openAI = OpenAI()

    response = openAI.images.generate(
        # 2026.05.12부로 dall-e-2, dall-e-3 서비스 종료
        # model="gpt-image-1.5"
        model="gpt-image-1-mini",
        prompt=f"{content}\n위 내용을 잘 설명할 수 있는 60년대 포스터 이미지 생성",
        size="1024x1024",
        n=1
    )

    image_base64 = response.data[0].b64_json
    image_bytes = base64.b64decode(image_base64)
    
    save_dir = "images"
    
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"image_{timestamp}.png"
    filepath = os.path.join(save_dir, filename)
    
    with open(filepath, "wb") as f:
        f.write(image_bytes)
    
    print(f"이미지가 성공적으로 저장되었습니다: {filepath}")
```

```python
def insert(content):
    print(f"INSERT DONE\n{content}\n")
    return content
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.runnables import RunnableLambda

llm = ChatOpenAI(
    temperature=0.2,
    model_name="gpt-4.1-nano"
)

prompt_template = PromptTemplate.from_template(
    "제목: {title}, 제목을 보고 내용 5문장을 제작하라. 앞뒤에 인사말이나 부연 설명을 붙이지마라."
)

chain = (
    prompt_template
    | llm
    | StrOutputParser()
    | RunnableLambda(insert)
    | RunnableLambda(generate_image_from_text)
)

chain.invoke("지구 온난화를 막는 방법")
```

### sora-2
- AI 영상 제작, 돈 먹는 하마

```python
# https://developers.openai.com/api/docs/guides/video-generation
# 비디오 생성 함수 정의
def generate_video_from_text(content):
    client = OpenAI()
    
    print(f"--- 텍스트를 기반으로 영상을 생성합니다 ---")
    
    # Sora 또는 지원 모델을 사용하여 이미지 기반 영상 생성 요청
    video_response = client.videos.create(
        model="sora-2",
        prompt=f"{content}",
        size="1280x720"
    )
    
    return {
        "text": content,
        "video_response": video_response # 최종 영상 주소
    }
```
